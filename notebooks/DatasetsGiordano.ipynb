# 1. Dataset: ONU Sustainability of Countries Development
> #### ONU Indexes UNSDG "United Nations Sustainable Development Group" (or simply SDG)
> #### https://www.kaggle.com/datasets/vittoriogiatti/unsdg-united-nations-sustainable-development-group

```python
df1 = pd.read_csv("datasets/unsdg_2002_2021.csv")
df1
```

```python
df_ld = df1.pivot_table(index='country', aggfunc='sum', values='land_degraded_over_total_land_area')
df_ld = df_ld.sort_values(by='land_degraded_over_total_land_area', ascending=False)
df_ld.reset_index(inplace=True)
```

```python
df_ld.head(11)
```

> En este dataset me parece interesante ver las relaciones entre nivel de desarrollo del país (level_of_development) y variables como tasa de mortalidad (mortality_rate_perc), territorio degradado sobre el total del territorio (land_degraded_over_total_land_area), en el contexto del uso de energías fosiles y renovables. 
> * Un indiciador significativo a nivel local es que encontramos a Argentina a un paso de entrar en el top 10 de países con mayor superficie de territorio degradado.

> Como pregunta sería interesante analizar qué aspectos se correlacionan con este indicador, buscando puntos en común entre los 10 países que conforman este top 10.

# 2. Dataset: Cause of Deaths around the World (Historical Data)
> #### This Dataset contains Historical Data on Causes of Death around the World
> #### https://www.kaggle.com/datasets/iamsouravbanerjee/cause-of-deaths-around-the-world

```python
df2 = pd.read_csv("datasets/cause_of_deaths.csv", index_col=[0])
df2
```

```python
# Se seleccionan datos de America del Sur
df2.reset_index(inplace=True)
stringlist = ['ARG', 'BOL', 'BRA', 'CHL', 'COL', 'ECU', 'GUY', 'PRY', 'PER', 'SUR', 'URY', 'VEN']
df2 = df2[df2['Code'].str.contains('|'.join(stringlist))]
df2 = df2.set_index("Code")
df2.head(20)
```

```python
df_iv = df2.pivot_table(index='Code', aggfunc='sum', values='Interpersonal Violence')
df_iv = df_iv.sort_values(by='Interpersonal Violence', ascending=False)
df_iv
```

> En este dataset encontramos diversas causas de muerte, que en principio no tendrían relación, ya que van desde aspectos psico-sociales hasta aspectos hereditarios o accidentales.
> Resultaría interesante analizar los datos agrupando países según algún criterio, como por ejemplo "países de América del Sur", como en la tabla presentada, y a su vez analizar los cambios a través del tiempo.

> Por otro lado, algunas de las "causas" podrían pensarse relacionadas a niveles de desarrollo de los países, como por ejemplo la muerte por malaria. Sería interesante buscar correlaciones entre su descenso/aumento y el descenso/aumento de otras causas.

# 3. Dataset: Electricity from 200 Geographies Dataset
> #### Yearly electricity generation, capacity, emissions and demand data
> #### https://www.kaggle.com/datasets/michaelbryantds/electricity-data-for-200-geographies

```python
df3 = pd.read_csv("datasets/yearly_full_release_long_format.csv", index_col=[0])
df3
```

```python
df_vvy = df3.pivot_table(index=['Country code','Variable'], aggfunc='sum', values=['Value', 'YoY absolute change'])
df_vvy = df_vvy.sort_values(by='YoY absolute change', ascending=False)
df_vvy.head(20)
```

```python
df_vvy.reset_index(inplace=True)
stringlist = ['USA', 'CHN', 'IND']
df3_uc = df_vvy[df_vvy['Country code'].str.contains('|'.join(stringlist))]
df3_uc = df3_uc.set_index("Variable")
df3_uc = df3_uc.sort_values(by='YoY absolute change', ascending=False)
df3_uc
```

> En este dataset se observan datos respecto a la generación de energía a nivel mundial, y cómo fluctúan los esfuerzos por cambiar los tipos de combustibles año a año.
